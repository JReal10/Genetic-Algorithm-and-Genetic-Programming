```python
import random
import numpy as np
import time
import math 
from functools import reduce

target = ["4","B","5","6","D",".","A","F"]
HexaDecimal = ["1","2","3","4","5","6","7","8","9","A","B","C","D","E","F","."]


## Define weights (w_i) and values (v_i) for each box and the maximum weight that can be carried.
N = len(target)

## Define GA hyper-parameters
p_xo  = 0.7
p_mut = 0.02
n_generations = 50
n_population  = 1000
n_tournament  = 5

## counter for ftness function evaluation 
count_eval = 0

## Measure runtime of the algortihm: start time
time_start = time.time()

def arrayToString(a):
  
  string = ''.join(a)

  return string

## create population as list of individuals = tuple(bitstring, fitness)
# parameter n = populatin size
def initialize_population(n):
    population = []
    for i in range(n):
        # create array of random bits
        chromosome = np.random.choice(HexaDecimal, N)
        population.append((chromosome,fitness(chromosome)))
    return population

def crossover():
    xo_point = np.random.randint(1, N-1)
    chromosome = np.hstack([population[tournament()][0][0:xo_point+1], 
                          population[tournament()][0][xo_point:-1]])
    idx = tournament(n_tournament, True)
    population[idx] = (chromosome, fitness(chromosome))
    return idx

def mutation(idx):
    r = (np.random.rand(N) <= p_mut)
    genotype = population[idx][0]
    for i in range(len(genotype)):
        if r[i] > 0:
            genotype[i] = not genotype[i]
    population[idx] = (genotype, fitness(genotype))
    
    
# (INVerse) tournament selection
def tournament(n=n_tournament, inv=False):
    idx = np.random.randint(0, len(population), n)
    f = np.array([population[i][1] for i in idx])
    i = f.argmin() if inv else f.argmax()
    return idx[i]

def fitness(chromosome):
    global count_eval
    count_eval += 1
    fitness = 0
    
    # compute weight of genotype
    for i in range(N):
      if chromosome[i] == target[i]:
        fitness += i ** 2

    return fitness


## generate random population
population = initialize_population(n_population);

a = np.array([population[i][1] for i in range(len(population))])

## GA loop
for g in range(n_generations):
    # for each individual...
    for i in range(len(population)):
        # XO or cloning?
        if np.random.random() <= p_xo:
            # XO operator and return index of individual
            idx = crossover()
        else:
            # Cloning operator (select individual)
            idx = tournament()  
        # mutation operator
        mutation(idx)

time_end = time.time()

f = np.array([population[i][1] for i in range(len(population))])
minFitness = reduce(min,f)
maxFitness = reduce(max, f)

for i in range(len(population)):
  if population[i][1] == maxFitness:
    solution = population[i][0]

print("Number of evaluations: ", str(count_eval))
print("Expected solution:", arrayToString(target))
print("Optimal solution:", arrayToString(solution))
print("Accuracy: ", round(maxFitness/fitness(target) * 100), "%")
print("Time in seconds:", time_end - time_start)
```
